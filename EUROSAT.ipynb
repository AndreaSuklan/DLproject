```python
from matplotlib import pyplot as plt
```

```python
from datasets import load_dataset

ds = load_dataset("blanchon/EuroSAT_RGB")
```

```python
ds
```

```python
# Sample 1000 examples, shuffle them
sample = ds["train"].shuffle(seed=42).select(range(1000))

# Get class names if available (depends on the dataset)
label_names = ds["train"].features["label"].names  # adjust key if different

# Loop through and plot max_imgs samples
max_imgs = 3
for i in range(max_imgs):
    img = sample[i]['image']
    print(f"Type of img:{type(img)}")
    label_idx = sample[i]['label']
    label_name = label_names[label_idx]
    
    plt.imshow(img)
    plt.title(f"Label: {label_idx} - {label_name}")
    plt.axis('off')
    plt.show()
```
